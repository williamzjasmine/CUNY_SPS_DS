```python
import random
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
```

```python
results = []
flip = random.randint(0, 1)

for i in range(100000):
    flip = 1
    earnings = 2
    while flip == 1:
        flip = random.randint(0, 1)
        if flip == 1:
            earnings = earnings * 2
    results.append(earnings)
```

```python
results_agg = Counter(results)
results_df = pd.DataFrame(zip(results_agg.keys(), results_agg.values()), columns = ['earnings', 'count'])
results_df = results_df.sort_values(by='earnings')

plt.figure(figsize=(10,8))
plt.bar(np.arange(0, results_df.shape[0]), results_df['count'])
plt.xticks(np.arange(0, results_df.shape[0]), results_df.earnings, rotation=45)
plt.xlabel('Earnings', fontweight='bold')
plt.title('Distribution of Earnings Won (total trials = 100,000)', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.grid()
plt.show()
```
